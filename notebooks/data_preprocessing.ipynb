```python
cols = ['Global_active_power','Global_reactive_power','Voltage',
        'Global_intensity','Sub_metering_1','Sub_metering_2','Sub_metering_3']

df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df=df.sort_index()
```

```python
df.head()
```

```python
power_hourly = df['Global_active_power'].resample('H').mean()
```

```python
power_hourly = power_hourly.dropna()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
power_scaled = scaler.fit_transform(power_hourly.values.reshape(-1,1))
```

```python
import pandas as pd
power_scaled = pd.DataFrame(power_scaled, index=power_hourly.index, columns=['scaled_power'])
```

```python
split_index = int(len(power_scaled) * 0.8)
```

```python
train = power_scaled.iloc[:split_index]
test = power_scaled.iloc[split_index:]
```

```python
print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

```python
power_scaled.to_csv("../data/final_prepared_data.csv")
```
