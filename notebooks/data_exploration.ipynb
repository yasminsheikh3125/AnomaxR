```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/household_power_consumption_clean.csv")

df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
df.set_index('Datetime', inplace=True)

df['Global_active_power'] = pd.to_numeric(df['Global_active_power'], errors='coerce')
df = df.dropna()
```

```python
daily = df['Global_active_power'].resample('D').mean()

plt.figure(figsize=(12,4))
plt.plot(daily)
plt.title("Daily Average Electricity Consumption")
plt.xlabel("Date")
plt.ylabel("Power")
plt.show()
```

```python
plt.hist(df['Global_active_power'], bins=50)
plt.title("Distribution of Electricity Consumption")
plt.xlabel("Power Usage")
plt.ylabel("Frequency")
plt.show()
```
